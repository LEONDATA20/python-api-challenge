```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
#print(weather_api_key)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
#print(cities)

list_of_tuples = list(lat_lngs)
type(list_of_tuples)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?"

units = "imperial"

#query_url = f"{url}appid={weather_api_key}&q="
query_url = f"{url}appid={weather_api_key}&units={units}&q="

temperatures = []
count = 0
_set = 1

print("Beginning Data Retrieval ")
print("--------------------------------")
for city in cities:
    
    count += 1
    if count >= 50:
        _set += 1
        count = 0
    response = requests.get(query_url + city).json()

#     pprint(response)

    try:
        print(f"Processing Record {count} of Set {_set} | {response['name']}")
        #print(f"Temperature is {response['main']['temp']}")

    except: 
       
        print("City not found. Skipping...")
    pass
                          
print("--------------------------------")
print("Data Retrieval Complete ")
print("--------------------------------")



```

```python

print(requests.get(query_url + city).status_code)
from pprint import pprint
pprint(response)

```

```python
temperatures = []
lats=[]
Countries=[]
Dates=[]
Humidities=[]
Cloudinesses=[]
lngs=[]
max_temps=[]
wind_speeds=[]
Citys=[]
for city in cities:  
    response = requests.get(query_url + city).json()
    
    
    try:
        temperature = response['main']['temp']
        country=response['sys']['country']
        date=response['dt']
        humidity=response['main']['humidity']
        cloudiness=response['clouds']['all']
        lat=response['coord']['lat']
        lng=response['coord']['lon']
        max_temp=response['main']['temp_max']
        wind_speed=response['wind']['speed']
        _city=response['name']
    
        temperatures.append(temperature)
        lats.append(lat)
        Countries.append(country)
        Dates.append(date)
        Humidities.append(humidity)
        Cloudinesses.append(cloudiness)
        lngs.append(lng)
        max_temps.append(max_temp)
        wind_speeds.append(wind_speed)
        Citys.append(_city)
        #print(Cloudinesses)
        print(temperatures)
        #print(Dates)
    except:
        print("N/A")
        pass
     
       
```

```python
print(len(temperatures))
print(len(lats))
print(len(Countries))
print(len(Dates))
print(len(Humidities))
print(len(Cloudinesses))
print(len(lngs))
print(len(max_temps))
print(len(wind_speeds))
print(len(Citys))
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_dict = {
    "City": Citys,
    "Cloudiness": Cloudinesses,
    "Country": Countries,
    "Date":Dates,
    "Humidity":Humidities,
    "Lat":lats,
    "Lng":lngs,
    "Max Temp":max_temps,
    "Wind Speed":wind_speeds
     }
weather_data = pd.DataFrame(weather_dict)
weather_data
weather_data.dropna()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = northern_df["Lat"]
y_values = northern_df["Max Temp"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
```

```python
plt.figure(figsize=(8,6))
plt.scatter(northern_df["Lat"], northern_df["Max Temp"], marker="o",facecolors="blue", edgecolors="black")
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5,-10),fontsize=15,color="red")            
plt.xlabel('Latitude')
plt.ylabel('Max Temp')
plt.title("Northern Hemisphere - Max Temp vs. Latitude")

# plt.figure(figsize=(8,6))
print(f"The r-squared is: {rvalue}")
# f = gcf;
# exportgraphics(f,'barchart.png','Resolution',300)
plt.savefig("Latitude vs. Humidity.png")
plt.show()
```

```python
#The r-value is quite high which means the value is quite predictable. The higher the latitude, the lower the max temp.


```
